```python
import pandas  as pd # data loading and manipulation
import numpy as np # mathematical calculatiion
import matplotlib.pyplot as plt # visualization
import seaborn as sns # visualization statistical
```

```python
# read the contents of a CSV file and print the first five rows
df=pd.read_csv("Global YouTube Statistics.csv",encoding='latin-1')
df.head()
```

```python
 #checking the missing value in each column
 df.isnull().sum()
```

```python
#checking the duplicate value
df.duplicated().sum()
```

```python
df.columns.tolist() #arrange feature in list form
```

```python
#drop the irrelavant columns
columns_to_drop = ['Gross tertiary education enrollment (%)', 'Population', 'Unemployment rate', 'Urban_population', 'Latitude', 'Longitude']

df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
#top youtube channel in terms of subscribers
plt.figure(figsize=(15,8))
plt.bar(height = 'subscribers', x = 'Youtuber' , data= df.head(15))
plt.xticks(rotation = 90)
plt.title("Top 15 youtuber in terms of subscribers ")
plt.show()
```

#Top Youtube Channel in terms of views

```python
df2 = df.copy()
df2.head
```

```python
#we sorting our dataset in terms of video views
df2 = df2.sort_values('video views' , ascending = False)
df2.head
```

```python
#on the basis of views this are the top channel types.
df2.head(100)['channel_type'].value_counts()
```

```python
df.head(30)['Country'].value_counts()
```

```python
#top youtube channel in terms of video view
sns.set_style('darkgrid')

plt.figure(figsize=(15,8))
ax = sns.barplot(y = 'video views', x = 'Youtuber' , data= df2.head(10) , palette='magma')
ax.set_yticklabels([f'{label/1000000}M' for label in ax.get_yticks()])
plt.xticks(rotation = 90)
plt.title("Top 15 youtuber in terms of video views  ")
plt.show()
```

```python
#top 5 channel categories in terms of subscribers
top_categories_subscribers = df.groupby('category')['subscribers'].sum().reset_index()


top_5_categories_subscribers = top_categories_subscribers.sort_values(by='subscribers', ascending=False).head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='subscribers', y='category', data=top_5_categories_subscribers, palette='viridis')
plt.title('Top 5 Channel Categories in Terms of Subscribers')
plt.xlabel('subscribers')
plt.ylabel('category')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 Countries by Numbers of Views
top_countries_views = df.groupby('Country')['video views'].sum().reset_index()

top_10_countries_views = top_countries_views.sort_values(by='video views', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries_views['Country'], top_10_countries_views['video views'], color='skyblue')
plt.title('Top 10 Countries by Number of Views')
plt.xlabel('country')
plt.ylabel('Number of Views')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# box plot of all numerical column
int_cols = df.select_dtypes(include=int)
plt.figure(figsize=(28,10))
# i referencing subplots , j is referencing column names
for i,j in zip(range(1,10), int_cols):
  plt.subplot(3,3,i)
  plt.subplots_adjust(right=0.9,top=1.8)
  sns.boxplot(data=df,x=j)
  plt.title("box plot of {}".format(j))
```

```python


top_5_highest_earnings = df.sort_values(by='highest_yearly_earnings', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_5_highest_earnings['Youtuber'], top_5_highest_earnings['highest_yearly_earnings'], color='purple')
plt.title('Top 5 Youtubers by Highest Yearly Earnings')
plt.xlabel('Youtuber')
plt.ylabel('Highest Yearly Earnings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

created_year_counts = df['created_year'].value_counts().reset_index().rename(columns={'index': 'Year', 'created_year': 'Count'})

created_year_counts = created_year_counts.sort_values(by='Year')
plt.figure(figsize=(10, 6))
plt.bar(created_year_counts['Year'], created_year_counts['Count'], color='orange')
plt.title('Number of YouTubers Created Each Year')
plt.xlabel('Year')
plt.ylabel('Number of YouTubers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

df_after_2010 = df[df['created_year'] > 2010]

top_5_highest_earnings = df_after_2010.sort_values(by='highest_yearly_earnings', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_highest_earnings['Youtuber'], top_5_highest_earnings['highest_yearly_earnings'], color='purple')
plt.title('Top 5 Youtubers (created after 2010) by Highest Yearly Earnings')
plt.xlabel('Youtuber')
plt.ylabel('Highest Yearly Earnings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

top_5_uploads = df.sort_values(by='uploads', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_5_uploads['Youtuber'], top_5_uploads['uploads'], color='green')
plt.title('Top 5 Youtubers by Number of Uploads')
plt.xlabel('Youtuber')
plt.ylabel('Number of Uploads')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

bottom_5_subscribers = df.sort_values(by='subscribers', ascending=True).head(5)
plt.figure(figsize=(10, 6))
plt.bar(bottom_5_subscribers['Youtuber'], bottom_5_subscribers['subscribers'], color='blue')
plt.title('Bottom 5 Youtubers by Subscribers')
plt.xlabel('Youtuber')
plt.ylabel('Subscribers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

top_5_subscribers_last_30_days = df.sort_values(by='subscribers_for_last_30_days', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_5_subscribers_last_30_days['Youtuber'], top_5_subscribers_last_30_days['subscribers_for_last_30_days'], color='orange')
plt.title('Top 5 Youtubers by Subscribers Gained in Last 30 Days')
plt.xlabel('Youtuber')
plt.ylabel('Subscribers Gained (Last 30 Days)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

top_youtubers = df.sort_values(by='subscribers', ascending=False).head(10)
created_month_counts = top_youtubers['created_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
created_month_counts.plot(kind='bar', color='skyblue')
plt.title('Top Youtubers Created by Month')
plt.xlabel('Month')
plt.ylabel('Number of Youtubers')
plt.xticks(rotation=0)
plt.show()
```
